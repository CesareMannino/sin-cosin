```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
import numpy as np

# fig = plt.figure()
# ax = plt.axes()

x = np.linspace(-360,360, num=180)
#np.deg2rad function as argument takes the position of x 
x = np.deg2rad(x)


plt.plot(x, np.sin(x + 3.14), color = 'red', label = 'Sin(x)')
#a list is created indexing where the dots need to be placed
#matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=<deprecated parameter>, edgecolors=None, \*, plotnonfinite=False, data=None, \*\*kwargs)[source]
plt.scatter([-np.deg2rad(270),-np.deg2rad(270),np.deg2rad(90),np.deg2rad(90)],[-1,1,-1,1],s = 50,c='black',marker='+')
plt.plot(x, np.cos(x - 1.6), color = 'blue', label = 'Cos(x)')
plt.scatter([-np.deg2rad(90),-np.deg2rad(90),np.deg2rad(270),np.deg2rad(270)],[-1,1,-1,1],s = 50,c='orange',marker='o')
plt.title('Sin & Cosin curve 180 deg difference')
plt.xlabel('Angle(rad)')
plt.ylabel('Sin(x): blue \n Cos(x) : orange')

plt.legend(loc = 3)
```
